```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Membaca file CSV
data = pd.read_csv('/content/UAS/kebutuhan_per_bulan_2.csv')
```

```python
data
```

```python
# Menampilkan scatter plot
plt.figure(figsize=(12, 6))  # Mengatur ukuran figure
plt.scatter(data['Propinsi'], data['tahun'], s=50, c='blue', alpha=0.7)

# Menambahkan label pada setiap titik data
for i in range(len(data)):
    plt.annotate(data['kebutuhan'][i], (data['Propinsi'][i], data['tahun'][i]), ha='center', va='bottom', fontsize=8)

# Menambahkan judul dan label sumbu
plt.title('PLOT 1 - Kebutuhan Per Tahun Berdasarkan Provinsi', fontsize=14)
plt.xlabel('Propinsi', fontsize=12)
plt.ylabel('Tahun', fontsize=12)

# Mengatur rotasi dan ukuran font label sumbu X
plt.xticks(rotation=90, fontsize=8)

# Menambahkan grid
plt.grid(True, linestyle='--', alpha=0.5)

# Menampilkan scatter plot
plt.tight_layout()
plt.show()
```

```python

data_filtered_jabar = data[data['Propinsi'] == 'BALI']

plt.scatter(data_filtered_jabar['tahun'], data_filtered_jabar['kebutuhan'], c=data_filtered_jabar.index, cmap='rainbow', s=50, alpha=0.7)

plt.title('PLOT 2 - Kebutuhan per Tahun di Provinsi BALI')
plt.xlabel('Tahun')
plt.ylabel('Kebutuhan')

cbar = plt.colorbar()
cbar.set_label('Data Index')

plt.show()
```

```python
# Membaca file CSV
df_transaksi = pd.read_csv("/content/UAS/transaksi2.csv", names=['date', 'id_product', 'quantity', 'id_penjualan'])
df_transaksi
```

```python
df_transaksi['date'] = pd.to_datetime(df_transaksi['date'])
df_transaksi['year'] = df_transaksi['date'].dt.year
df_transaksi['month'] = df_transaksi['date'].dt.month_name()
df_clean = df_transaksi.loc[df_transaksi['year'] == 2017]
# Melakukan pengelompokan dan penghitungan
plt.title('PLOT 3 - Data Transaksi Tahun 2017')
df_count = df_clean.groupby('month').count()
# Menampilkan scatter plot
plt.scatter(df_count['date'], df_count.index)
```
